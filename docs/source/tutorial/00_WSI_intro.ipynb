# Whole slide images  Intro

Whole slide imaging (WSI) allows for the digital scanning of tissue slides at gigapixel resolution. These images are commonly referred to as whole slide images (WSIs) and are used in routine pathology diagnostic. Unlike in traditional microscopy, WSIs provide a comprehensive digital representation of tissue section of up to **75 mm x 25 mm**, but typically large sections covering an area of **20-40 mm x 15-25 mm**. This enables more efficient, scalable, and reproducible analyses, making them invaluable in both research and clinical settings.

Typical WSIs file formats are:

- **.SVS (Aperio):** Widely used in both research and clinical settings, this format is compatible with many image analysis tools.
- **.NDPI (Hamamatsu):** Commonly used in academic and research institutions.
- **.TIFF:** A versatile open source format that can store  large images and is supported by many imaging tools

WSIs are often stored as **pyramidal images** with multiple levels of resolutions, with the highest resolution being level 0 to lower, downsampled versions. Another important information about you WSIs is the **microns per pixel** (mpp) they are scanned at, being the physical size of a pixel in microns, which is crucial for scaling and quantitative analysis. When you move across WSIs levels, the images will be downsampled by a given factor, which is saved in level_downsample.
